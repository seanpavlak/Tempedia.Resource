```python
import glob
import re
import requests
import pandas as pd
```

```python
print('https://www.temtem-france.com/tempedie')
print('')
```

```python
luma_images = glob.glob('..//images//luma//*.png')
luma_gifs = glob.glob('..//images//luma//*.gif')
standard_gifs = glob.glob('..//images//standard//*.gif')
standard_images = glob.glob('..//images//standard//*.png')
```

```python
request_json = requests.get('https://temtem-api.mael.tech/api/temtems').json()
```

```python
api_response_ids = []
luma_image_ids = []
luma_gif_ids = []
standard_image_ids = []
standard_gif_ids = []
```

```python
for item in request_json:
    api_response_ids.append([item.get('number'), item.get('name')])
    
for item in luma_images:
    luma_image_ids.append(int(re.findall(r'\d{1,9}',item)[0]))
    
for item in luma_gifs:
    luma_gif_ids.append(int(re.findall(r'\d{1,9}',item)[0]))

for item in standard_images:
    standard_image_ids.append(int(re.findall(r'\d{1,9}',item)[0]))

for item in standard_gifs:
    standard_gif_ids.append(int(re.findall(r'\d{1,9}',item)[0]))
```

```python
api_response_ids.sort()
luma_image_ids.sort()
luma_gif_ids.sort()
standard_image_ids.sort()
standard_gif_ids.sort()
```

```python
df = pd.DataFrame(api_response_ids, columns =['ID', 'Name'])
df.set_index('ID', inplace = True)
df['Standard_Image_Check'] = False
df['Standard_GIF_Check'] = False
df['Luma_Image_Check'] = False
df['Luma_GIF_Check'] = False
```

```python
for item in standard_image_ids:
    df.at[item,'Standard_Image_Check'] = True
    
for item in standard_gif_ids:
    df.at[item,'Standard_GIF_Check'] = True
    
for item in luma_image_ids:
    df.at[item,'Luma_Image_Check'] = True
    
for item in luma_gif_ids:
    df.at[item,'Luma_GIF_Check'] = True
```

```python
# Missing Standard Images
df_Standard_Image_Check = df[df["Standard_Image_Check"] == False]
print(f'Missing Standard {len(df_Standard_Image_Check)} Images')
df_Standard_Image_Check[:]
```

```python
# Missing Standard Gifs
df_Standard_GIF_Check = df[df["Standard_GIF_Check"] == False]
print(f'Missing Standard {len(df_Standard_GIF_Check)} GIFs')
df_Standard_GIF_Check[:]
```

```python
# Missing Luma Images
df_Luma_Image_Check = df[df["Luma_Image_Check"] == False]
print(f'Missing Luma {len(df_Luma_Image_Check)} Images')
df_Luma_Image_Check[:]
```

```python
# Missing Luma Gifs
df_Luma_GIF_Check = df[df["Luma_GIF_Check"] == False]
print(f'Missing Luma {len(df_Luma_GIF_Check)} GIFs')
df_Luma_GIF_Check[:]
```
